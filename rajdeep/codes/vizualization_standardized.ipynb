```python
import pandas as pd
import torch
from scipy import stats
```

```python
def normalize_sequence(sequence):
# 	sequence_max = max(sequence)
# 	sequence_min = min(sequence)
# 	norm = [(float(i)-sequence_min)/(sequence_max-sequence_min) for i in sequence]
	stan = stats.zscore(sequence)
	return stan.tolist()
	return norm
```
